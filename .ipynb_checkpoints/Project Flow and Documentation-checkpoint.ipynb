<h2 style="color:MidnightBlue">Data Cleaning</h2> 

After the collection of the images, what we are gonna do of it? 
A simple question, how we basically identify a person, first, we see the face of a person, but every person have different face features, but the most prominent face feature is our eyes.
So, we first see into the eyes of the person to identify him.

In case of face detection using python, we keep only those images which is clearly visible and the eyes and face are not obstructed, rest of the images will be discarded. 


For face detection and detecting the eyes we will use openCV, and for the specific detection, we will use hars cascade in python.

In the overall data cleaning process, the Raw Images will be passed through the haar cascade algorithm which will identify the faces with two eyes, the outcome will be processed under a manual data cleaning in case any image is not of the part of our classes or some unnecessary image, then the further product will be gone through the wavelet transfrom for the feature engineering, at the last these final images will be used to train our model and this model will be exported to a file, and then a python flask server will run around it which will serve to our website as our final project
